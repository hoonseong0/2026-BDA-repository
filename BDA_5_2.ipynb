```python
courses = pd.Series(["Spark", "PySpark", "Hadoop"], name='courses')
courses

fees = pd.Series([250000, 300000, 200000], name='fees')
fees

df = pd.DataFrame({courses.name: courses, fees.name: fees})
df
```

```python
courses = pd.Series(["Spark", "PySpark", "Hadoop"], name='courses')
fees = pd.Series([250000, 300000, 200000], name='fees')

df = pd.concat([courses, fees], axis=1)
df
```
